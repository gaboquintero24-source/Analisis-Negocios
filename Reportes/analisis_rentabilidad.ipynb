# Cálculo de Rentabilidad de Exportación
En este script calcularemos el margen de ganancia de un producto FOB.

```python
# Definimos los costos y precios (Datos de entrada)
nombre_producto = "Café Especial Colombiano"
costo_produccion = 15.50  # Dólares por libra
precio_fob = 8.00        # Dólares por libra
cantidad_exportada = 500  # Libras

print("Datos cargados exitosamente para:", nombre_producto)
```

```python
# Cálculos
ingreso_total = precio_fob * cantidad_exportada
costo_total = costo_produccion * cantidad_exportada
ganancia_neta = ingreso_total - costo_total

# Fórmula de margen: (Ganancia / Ingreso) * 100
margen_porcentaje = (ganancia_neta / ingreso_total) * 100

print(f"Ingreso Total: ${ingreso_total}")
print(f"Ganancia Neta: ${ganancia_neta}")
print(f"Margen de Ganancia: {margen_porcentaje:.2f}%")
```
