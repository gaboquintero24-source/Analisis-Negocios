```python
import pandas as pd

# Cargamos el archivo que creamos en la carpeta Datos
# El '..' significa "sube una carpeta y luego entra a Datos"
try:
    df = pd.read_csv('../Datos/productos_exportacion.csv')
    print("¡Tabla cargada con éxito!")
except FileNotFoundError:
    print("Error: No encuentro el archivo. Verifica que la carpeta 'Datos' exista y el nombre sea correcto.")

# Mostramos los primeros datos
if 'df' in locals():
    display(df.head())
```

```python
# Creamos una columna nueva calculada automáticamente
df['ganancia_unidad'] = df['precio_fob'] - df['costo_unidad']
df['total_venta'] = df['precio_fob'] * df['cantidad']

# Calculamos el margen porcentual y lo redondeamos a 2 cifras
# Usamos .round(2) al final del cálculo
df['margen_%'] = ((df['ganancia_unidad'] / df['precio_fob']) * 100).round(2)

# Mostramos el resultado
display(df)
```

```python
# Filtramos la tabla
productos_rentables = df[df['margen_%'] > 40]

print("Productos con alta rentabilidad (Mayor al 40%):")
display(productos_rentables)
```
